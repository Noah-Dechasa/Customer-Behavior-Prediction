```python
from sklearn.model_selection import train_test_split

X = behavior[['Age', 'Purchase Amount (USD)', 'Frequency of Purchases', 'Review Rating', 'Discount Applied']]  
y = behavior['Subscription Status']  

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
encoder = LabelEncoder()

# Apply Label Encoding to 'Frequency of Purchases'
X_train['Frequency of Purchases'] = encoder.fit_transform(X_train['Frequency of Purchases'])
X_test['Frequency of Purchases'] = encoder.transform(X_test['Frequency of Purchases'])

# Check if the encoding was applied successfully
print(X_train['Frequency of Purchases'].head())

# Apply Label Encoding to 'Discount Applied'
X_train['Discount Applied'] = encoder.fit_transform(X_train['Discount Applied'])
X_test['Discount Applied'] = encoder.transform(X_test['Discount Applied'])

# Check if the encoding was applied successfully
print(X_train['Discount Applied'].head())

# Align the columns in X_train and X_test
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

# Check the final shape of the datasets
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create and fit the Logistic Regression model
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train, y_train)

# Predict on the test set
y_pred_log = log_model.predict(X_test)

# Evaluate the model
log_accuracy = accuracy_score(y_test, y_pred_log)
log_classification_report = classification_report(y_test, y_pred_log)
log_confusion_matrix = confusion_matrix(y_test, y_pred_log)

# Print results
print(f'Logistic Regression Accuracy: {log_accuracy:.4f}')
print('Classification Report (Logistic Regression):\n', log_classification_report)
print('Confusion Matrix (Logistic Regression):\n', log_confusion_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create and fit the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_classification_report = classification_report(y_test, y_pred_rf)
rf_confusion_matrix = confusion_matrix(y_test, y_pred_rf)

# Print results
print(f'Random Forest Accuracy: {rf_accuracy:.4f}')
print('Classification Report (Random Forest):\n', rf_classification_report)
print('Confusion Matrix (Random Forest):\n', rf_confusion_matrix)

```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get best parameters
print(f"Best Parameters from GridSearch: {grid_search.best_params_}")

# Evaluate the best model from GridSearch
best_rf_model = grid_search.best_estimator_
y_pred_best_rf = best_rf_model.predict(X_test)

# Evaluate the tuned model
best_rf_accuracy = accuracy_score(y_test, y_pred_best_rf)
best_rf_classification_report = classification_report(y_test, y_pred_best_rf)
best_rf_confusion_matrix = confusion_matrix(y_test, y_pred_best_rf)

# Print results
print(f'Tuned Random Forest Accuracy: {best_rf_accuracy:.4f}')
print('Classification Report (Tuned Random Forest):\n', best_rf_classification_report)
print('Confusion Matrix (Tuned Random Forest):\n', best_rf_confusion_matrix)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create and fit the Gradient Boosting model
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# Predict on the test set
y_pred_gb = gb_model.predict(X_test)

# Evaluate the model
gb_accuracy = accuracy_score(y_test, y_pred_gb)
gb_classification_report = classification_report(y_test, y_pred_gb)
gb_confusion_matrix = confusion_matrix(y_test, y_pred_gb)

# Print results
print(f'Gradient Boosting Accuracy: {gb_accuracy:.4f}')
print('Classification Report (Gradient Boosting):\n', gb_classification_report)
print('Confusion Matrix (Gradient Boosting):\n', gb_confusion_matrix)

```

```python
from sklearn.svm import SVC

# Create and fit the SVM model
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Predict on the test set
y_pred_svm = svm_model.predict(X_test)

# Evaluate the model
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_classification_report = classification_report(y_test, y_pred_svm)
svm_confusion_matrix = confusion_matrix(y_test, y_pred_svm)

# Print results
print(f'SVM Accuracy: {svm_accuracy:.4f}')
print('Classification Report (SVM):\n', svm_classification_report)
print('Confusion Matrix (SVM):\n', svm_confusion_matrix)
```

```python
# Comparing model accuracies
print(f"Logistic Regression Accuracy: {log_accuracy:.4f}")
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
print(f"Tuned Random Forest Accuracy: {best_rf_accuracy:.4f}")
print(f"Gradient Boosting Accuracy: {gb_accuracy:.4f}")
print(f"SVM Accuracy: {svm_accuracy:.4f}")

# Display Classification Reports for All Models
print("\nLogistic Regression Classification Report:")
print(log_classification_report)

print("\nRandom Forest Classification Report:")
print(rf_classification_report)

print("\nTuned Random Forest Classification Report:")
print(best_rf_classification_report)


print("\nGradient Boosting Classification Report:")
print(gb_classification_report)

print("\nSVM Classification Report:")
print(svm_classification_report)
```

Best Model Overall: Tuned Random Forest seems to be the best model, providing a good balance of precision and recall for both classes, with high accuracy.
Best Model for Predicting 'Yes': Gradient Boosting and Tuned Random Forest provide good recall for 'Yes', making them excellent for churn prediction or any task where predicting the positive class is important.
Worst Model: SVM is not suitable for this problem, as it completely fails to predict the 'Yes' class.

```python
# Select only the relevant columns: Discount Applied, Purchase Amount, Age
X_reduced = X_train[['Discount Applied', 'Purchase Amount (USD)', 'Age']]
X_test_reduced = X_test[['Discount Applied', 'Purchase Amount (USD)', 'Age']]

# Check the shape of the new dataset
print(X_reduced.shape, X_test_reduced.shape)

# Train a Logistic Regression model with the reduced dataset
log_model_reduced = LogisticRegression(max_iter=1000, random_state=42)
log_model_reduced.fit(X_reduced, y_train)

# Predict on the test set
y_pred_log_reduced = log_model_reduced.predict(X_test_reduced)

# Evaluate the model
log_accuracy_reduced = accuracy_score(y_test, y_pred_log_reduced)
log_classification_report_reduced = classification_report(y_test, y_pred_log_reduced)
log_confusion_matrix_reduced = confusion_matrix(y_test, y_pred_log_reduced)

# Train Random Forest on the reduced dataset
rf_model_reduced = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_reduced.fit(X_reduced, y_train)

# Predict and evaluate
y_pred_rf_reduced = rf_model_reduced.predict(X_test_reduced)

rf_accuracy_reduced = accuracy_score(y_test, y_pred_rf_reduced)
rf_classification_report_reduced = classification_report(y_test, y_pred_rf_reduced)
rf_confusion_matrix_reduced = confusion_matrix(y_test, y_pred_rf_reduced)

# Print results
print(f'Random Forest (Reduced Features) Accuracy: {rf_accuracy_reduced:.4f}')
print('Classification Report (Reduced Features) - Random Forest:\n', rf_classification_report_reduced)
print('Confusion Matrix (Reduced Features) - Random Forest:\n', rf_confusion_matrix_reduced)


# Print results
print(f'Logistic Regression (Reduced Features) Accuracy: {log_accuracy_reduced:.4f}')
print('Classification Report (Reduced Features) - Logistic Regression:\n', log_classification_report_reduced)
print('Confusion Matrix (Reduced Features) - Logistic Regression:\n', log_confusion_matrix_reduced)


# Comparing accuracy of the reduced model and the full model
print(f"Logistic Regression (Full Model) Accuracy: {log_accuracy:.4f}")
print(f"Logistic Regression (Reduced Features) Accuracy: {log_accuracy_reduced:.4f}")

print(f'Random Forest (Full model) Accuracy: {rf_accuracy:.4f}')
print(f'Random Forest (Reduced Features) Accuracy: {rf_accuracy_reduced:.4f}')
```

After evaluating multiple models, I was found that Discount Applied, Purchase Amount, and Age were the most important predictors of customer behavior. Logistic Regression performed equally well with both the full and reduced models, maintaining an accuracy of 82.99%, showing these three features are highly predictive. Random Forest showed a slight decrease in accuracy when reduced to just these features, but the performance was still strong.
